```
import matplotlib.pyplot as plt
import numpy as np
import os
import random
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator,array_to_img,img_to_array
%matplotlib inline
from matplotlib import pyplot as plt
import cv2
```

```
datagen = ImageDataGenerator(
     rotation_range=15)
```

```
imgs = os.listdir('/content/drive/My Drive/Colab Notebooks/tmp/test/potholesdslr/')
```

```
for img in imgs:
    img=cv2.imread('/content/drive/My Drive/Colab Notebooks/tmp/train/potholes/'+"//"+img)
    x = img_to_array(img)
    x = x.reshape((1,) + x.shape)  

    i = 0
    for batch in datagen.flow (x, batch_size=1, save_to_dir ='/content/drive/MyDrive/Colab Notebooks/tmp/rotationrange/potholes95', save_prefix ='rotation20', save_format='jpg'):
        i+=1
        if i>2:
            break
```
